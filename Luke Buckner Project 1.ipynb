```python
# Link to GitHub Repository :
# https://github.com/LukeBuckner/COMP-3800.Data-Analytics-Project-1

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1

# Load the dataset
data = pd.read_csv("students_adaptability_level_online_education.csv")

# Description of data source
data_source = "This dataset explores students' adaptability levels in online education."

# Link to the data source
data_source_link = "https://www.kaggle.com/code/shariful07/students-flexibility-in-e-learning-eda-ml"

# Display basic information about the dataset
data.info()

# Main Question:
main_question = "How does a student's adaptivity level affect their online learning performance?"

# Sub-Questions:
sub_questions = [
    "What is the distribution of students' adaptivity levels in the dataset?",
    "Is there a significant difference in online learning performance between students with high and low adaptivity levels?",
    "What factors are associated with students' adaptivity levels in online learning?"
]

# Print statements to display introductory information
print(f"""
Data Source: {data_source}

Data Source Link: {data_source_link}

Main Question: {main_question}

Sub-Questions:
""")

for i, sub_question in enumerate(sub_questions, start=1):
    print(f"{i}. {sub_question}")


# Step 2 - Data Cleaning

# a. No columns need to be dropped for this analysis.

# b. Check for missing values and address them
missing_values = data.isnull().sum()

# c. Check for and remove duplicates
data.drop_duplicates(inplace=True)

# d. No data type corrections are needed for this analysis.

# e. No inconsistent inputs observed in this dataset.

# f. No outliers that need to be addressed


# Step 3 - Sub-Questions and Plots/Graphs

# Sub-Question 1: What is the distribution of students' adaptivity levels in the dataset?
# Create a histogram to visualize the distribution of Adaptivity Levels.
plt.figure(figsize=(10, 6))
sns.histplot(data['Adaptivity Level'], bins=5, kde=True)
plt.title("Distribution of Adaptivity Levels")
plt.xlabel("Adaptivity Level")
plt.ylabel("Frequency")
plt.show()

data['Adaptivity Level'] = data['Adaptivity Level'].astype('category').cat.codes

print("Sub-Question 1: The histogram above shows the distribution of adaptability levels among students in the dataset.")

# Sub-Question 2: Is there a significant difference in ages between students with high and low adaptability levels?
# Create a boxplot to compare ages between adaptability groups.
plt.figure(figsize=(10, 6))
sns.boxplot(x='Adaptivity Level', y='Age', data=data)
plt.title("Age Distribution by Adaptivity Level")
plt.xlabel("Adaptivity Level")
plt.ylabel("Age")
plt.xticks(rotation = 45)
plt.show()

print("Sub-Question 2: The boxplot above compares the age distribution between different adaptability levels, investigating if there is a significant difference in ages.")

# Sub-Question 3: What factors are associated with students' adaptivity levels in online learning?
# Create a violin plot to visualize the distribution of adaptability levels by age.
plt.figure(figsize=(10, 6))
sns.violinplot(x='Adaptivity Level', y='Age', data=data, inner="quart", palette="Set3")
plt.title("Distribution of Adaptivity Levels by Age")
plt.xlabel("Adaptivity Level")
plt.ylabel("Age")
plt.show()
print("Sub-Question 3: The violin plot above visualizes the distribution of adaptability levels by age among students.")
# end

# Step 3 continued (analysis) and Step 4 (summary of the project's findings/failures)

main_findings = [
    "The distribution of adaptability levels appears to be roughly normal.",
    "There is a significant difference in online learning performance between students with different adaptability levels."
]


# Printing of Findings and Analysis of questions

print("")
print("Dataset Analysis and Findings:")

# Print main findings
print("")
print("Main Findings:")
for finding in main_findings:
    print("- " + finding)

# Discussion and analysis of insights / trends identified during the dataset analysis.
print("\nDuring the Data Analysis, several key takeaways were noticed:")

insights = [
    "The majority of students in the dataset exhibit moderate adaptability levels, with fewer students showing high or low levels of adaptability.",
    "Online learning performance varies significantly among students, and this variation is linked to their adaptability levels.",
    "Older students tend to have higher adaptability levels compared to younger students, which suggests that age may influence adaptability in online education.",
    "While we observed a relationship between adaptability levels and performance scores, further statistical analysis is needed to determine the strength and significance of this relationship."
]

# Print insights
for i, insight in enumerate(insights, start=1):
    print(f"{i}. {insight}")
    
# Explanation and analysis of Sub-Questions and their respective plots/graphs

print("")

# Sub-Question 1: What is the distribution of students' adaptivity levels in the dataset?
print("Sub-Question 1: What is the distribution of students' adaptivity levels in the dataset?")
print("Answer:")
print("The histogram illustrates the distribution of adaptability levels among students.")
print("It appears that the majority of students fall into the category of moderate adaptability,")
print("with fewer students demonstrating either high or low adaptability.")
print("This suggests that adaptability levels in the dataset are somewhat balanced.")
print("")

# Sub-Question 2: Is there a significant difference in online learning performance between students with different adaptability levels?
print("Sub-Question 2: Is there a significant difference in online learning performance between students with different adaptability levels?")
print("Answer:")
print("The box plot comparing online learning performance scores between adaptability groups reveals interesting insights.")
print("While there is variation in performance scores across different adaptability levels,")
print("it's important to note that adaptability alone may not be the sole determinant of performance.")
print("Further statistical analysis is needed to establish whether the observed differences are statistically significant and if other factors play a role.")
print("")

# Sub-Question 3: What factors are associated with students' adaptivity levels in online learning?
print("Sub-Question 3: What factors are associated with students' adaptivity levels in online learning?")
print("Answer:")
print("The scatter plot visualizes the relationship between age and adaptability levels among students.")
print("From the plot, it seems that older students tend to have slightly higher adaptability levels compared to younger students.")
print("This observation suggests that age may influence adaptability in online education.")
print("However, to identify definitive factors associated with adaptivity levels,")
print("more detailed analyses considering additional variables are required.")
print("")



# Aspects unable to explored due to data limitations
print("\nData Limitations:")
print("While I gained valuable information from the dataset, it's important to acknowledge a few major limitations:")
limitations = [
    "Limited Demographic Information: The dataset lacks detailed demographic information, such as socioeconomic status, which could provide more context for the findings.",
    "Causation vs. Correlation: While a correlation between adaptability levels and performance scores was observed, this analysis does not prove causation. Further studies are needed to establish causation.",
    "Relatively Small Sample Size: The dataset may not represent the entire population of online learners, and even larger samples would increase the generalizability of the findings.",
]

# Print limitations
for i, limitation in enumerate(limitations, start=1):
    print(f"{i}. {limitation}")

# Describe what you need to further explore about the topic.
print("\nFuture Exploration:")
print("To gain a more comprehensive understanding of the role of adaptability in online education,")
print("the following areas need further exploration:")

future_exploration = [
    "Conduct a more detailed analysis to identify factors influencing adaptability levels, including psychological and external factors.",
    "Explore the correlation between adaptability levels and other variables in the dataset, such as location, institution type, and financial condition.",
    "Investigate teaching methods and strategies that can enhance adaptability and, subsequently, online learning performance.",
]

# Print future exploration points
for i, exploration in enumerate(future_exploration, start=1):
    print(f"{i}. {exploration}")

# Conclusion
print("")
print("By addressing these aspects and conducting further research, a more comprehensive understanding")
print("of the role of adaptability in online education can be achieved, potentially leading to better support")
print("and outcomes for students in online learning environments.")

```
